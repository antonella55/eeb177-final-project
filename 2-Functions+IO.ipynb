# Function 1

**Create a Dictionary**
This function will take in the pbdb occurence data and sort through the specimens/lines and find all the specimens which have been identified to the species rank; this will be the value; the key is the specimen "occurence no"

```python
alligator_file=open("alligatoridae_pbdb_data.csv", "r")
alligator=alligator_file.readlines()[18:]
```

```python
#print(alligator)
import re
```

##### This bit of code allowed me to find which specimens were identified_rank=species
for line in alligator:
    if re.search(r"species", line):
        print(line)

```python
species_dict = {}

for line in alligator:
    items = line.split(",")
    occurence_no = items[0]
    species_name = items[5] #currently this column contains specimens identified to either the family, 
#genus or species rank; to retrieve info for only those identified to the species rank we write an if statement'''
        
    if re.search(r"species", line):
        species_dict[occurence_no] = species_name
```

# Function 2

**Make another dictionary**
This dictionary will contain as key the species name and as value the "max_ma" which is the maximum million years ago that this specimen represents

```python
max_ma_dict = {}

for line in alligator:
    items = line.split(",")
    max_ma = items[14]
    species_name = items[5] #again I sort for only occurence data identified to the species rank
    if re.search(r"species", line):
        max_ma_dict[species_name] = max_ma + str("mya")
    
```
